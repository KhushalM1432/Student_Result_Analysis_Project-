```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed Column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Change Weekly Study Hours Columns 

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5-Oct","5 - 10")
df.head()
```

# Gender Distribution

```python
sns.countplot(data = df, x = "Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data = df, x = "Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# From the above chart we have analysed that:
# The number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb)
plt.show()
```

```python
sns.heatmap(gb, annot = True)
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(gb, annot = True)
plt.title("Relationship Between Parent's Education and Student's Score")
plt.show()
```

# From the above chart we have concluded that the education of the parents have a good impact on their course 

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(gb1, annot = True)
plt.title("Relationship Between Parent's Marital Status and Student's Score")
plt.show()
```

# From the above chart we have concluded that there is no/negligible impact on the 
student's score due to their parent's marital status
